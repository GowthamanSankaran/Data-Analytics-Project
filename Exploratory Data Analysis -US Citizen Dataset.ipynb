#### Handling Null Values

```python
df.isna().sum()
```

```python
df['education-num']=df['education-num'].fillna(np.mean(df['education-num']))
```

```python
df['relationship']=df['relationship'].fillna("Undisclosed")
```

```python
df.isna().sum()
```

#### Handling Duplicate Values

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

#####                                                           Relationship Status of Data

```python
df['relationship'].value_counts()
```

```python
a=df['relationship'].value_counts()
fig=plt.figure()
ax=fig.add_axes([1,1,1.5,1.5])
ax.set_title('Relationship Status')
ax.pie(a,autopct='%1.0f%%',labels=['Husband',      
'Not-in-family','Undisclosed',    
'Unmarried',         
'Wife',             
'Other-relative' ],shadow=True)
```
