```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Importing Data for Analysis
df=pd.read_csv("c:\data.csv")
```

```python
df.head()
```

```python
#Deleting Irrelevant Column for Analysis
delcol=['Indicator Name','Indicator Code']
```

```python
df.drop( delcol, inplace=True ,axis=1)
```

```python
#Setting Index
df.set_index('Country Code')
```

```python
#Filling Empty Cells with Mean Value.
column=list(df)
for i in column:
    if i.isnumeric()==True:
        x=df[i].mean()
        df[i]=df[i].fillna(x)
```
